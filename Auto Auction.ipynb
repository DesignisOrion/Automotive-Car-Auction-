# Automotive Auction Data Project by DesignIsOrion

In this project, I was requested to explore the Automotive Auction Data provided by a client. The client wanted the dataset to be easy to understand as well as provide insight on the mercades benz listings that are available. 

Goal: 
- Fix the .csv and save a new modified version
- Find Pricing of Auto (mean, min/max)
- List Data of Mercades Benz Only


<img src="benz.png">



```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('cars.csv')
df.head(20)
```

```python
df.columns
```

```python
new_df= df.drop(['condition', 'country', 'Unnamed: 0', 'vin', 'title_status', 'mileage', 'lot', 'model'], axis=1)
new_df.columns= new_df.columns.str.upper()
new_df['BRAND']= new_df['BRAND'].str.upper()
new_df['COLOR']= new_df['COLOR'].str.upper()
new_df['STATE']= new_df['STATE'].str.upper()
new_df.reset_index(drop=True, inplace=True)
new_df.to_csv("modified.csv")
new_df
```

# Price

```python
# lets find the average PRICE on the list
new_df['PRICE'].mean()
```

```python
new_df['PRICE'].min()
```

```python
# highest priced car
new_df['PRICE'].max()
```

```python
# found the highest priced car.
new_df[new_df['PRICE']== 84900] 
```

```python
# Find out how many benzes are on the list.
benz_data = new_df[new_df['BRAND']== 'MERCEDES-BENZ']
benz_data.reset_index(drop=True, inplace=True)
benz_data.to_csv('benz_data.csv')
benz_data
```
